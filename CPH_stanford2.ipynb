# 3) Generate/estimate the survival curves for entire sample population



We create survival object using the ```Surv()``` function:

````r
surv_ob <- Surv( stanford_df$time, stanford_df$status )
class(surv_ob)
surv_ob[1:10]
````

````r
summary(surv_ob)
````

Apply the ```survfit``` function using the ```survival``` "object"

````r
km_model <- survfit( surv_ob ~ 1 )

rotate_y_labels <- T
title <- "KM curve"
plot( km_model,conf.int=T, xlab='month', ylab='% alive', main=title, mark.time=T, las=rotate_y_labels)
````

Time is categorized as low, mild, moderate, high

# Let's group by age categories

````r
stanford_df$age_category <- cut(stanford_df$age,
                       breaks=c(0, 18, 36, 60, 100),
                       labels=c('youth', 'adults', 'senior','elder') )
````

````r
hist(as.numeric( as.factor(stanford_df$age_category) ) )
````

````r
attach(stanford_df)
age_category
````

````r
km_model <- survfit( surv_ob ~ stanford_df$age_category )
rotate_y_labels <- T
title <- "KM curve"

legend_line_type=3

plot( km_model,conf.int=F, xlab='month', ylab='% alive', main=title, mark.time=T, las=rotate_y_labels, col=c('red','green','blue'), lty=legend_line_type)

label_grp1='youth'
label_grp2='adult'
label_grp3='senior'
boxtype=''
legend_line_width=3
legend_text_sz_rel2default = 1.2
legend(2000,.95, legend=c(label_grp1,label_grp2, label_grp3), col=c('red','blue','green'), cex=legend_text_sz_rel2default, bty=boxtype, lwd=legend_line_width, lty=legend_line_type )
````

# Q: Let's examine a binary independent variable 

````r
cox_model_binary <- coxph( surv_ob ~ final$over_40  )
summary( cox_model_binary )
````

# Q: Binary + categorical independent variables?

````r
surv_ob = Surv( final$time, final$status )
final$over_40 <- final$age > 40
cox_model_2vars <- coxph( surv_ob ~ final$over_40 + final$t5_category )
summary( cox_model_2vars )
````

# 4) Likelihood ration test to compare the two CPH models

````r
print( anova( cox_model_binary, cox_model_2vars, test='LRT') )
````

````r
print( anova( cox_model_2vars, cox_model_binary, test='LRT') )
````

# Q: Fit continuous values instead?

````r
cox_model_age <- coxph( surv_ob ~ final$age )
summary( cox_model_age )
````

````r
print( anova( cox_model_age, cox_model_binary, test='LRT') )
````

````r
print( anova( cox_model_age, cox_model_binary, test='LRT') )
````

# 5) Check Cox PH assumptions

- Linearity by martingale residual plot

````r
cox_cont_age_t5_model = coxph( surv_ob ~ final$age + final$t5 )

summary( cox_cont_age_t5_model )
````

````r
print( anova( cox_model_binary, cox_cont_age_t5_model, test='LRT') )
````

````r
pred <- predict( cox_cont_age_t5_model )
length(pred)
````

````r
resids = residuals(  cox_cont_age_t5_model, type='martingale' )

title='Martingale residual plot for checking linearity assumption'
plot( pred, resids, xlab ='Fitted values', ylab='Martingale residuals',main=title, las=1)

abline(h=0)
lines( smooth.spline( pred, resids), col='red')
````

# 6) Check for proportional hazard assuptions

````r
# cannot reject H0 that hazards are not proportional if P>0.05

cox.zph(  cox_cont_age_t5_model )
````

````r
# not proportional: confidence band falls at y=0 for about 2/3 of times only
plot( cox.zph(  cox_cont_age_t5_model ))
abline(h=0,col=2)
````

````r
# age variable

cox.zph(  cox_cont_age_t5_model )
plot( cox.zph(  cox_cont_age_t5_model )[1])
abline(h=0,col=2)
````

````r
# t5 variable
# 
# t5 meets PH assumption better than age as centered around 0 y=most times

plot( cox.zph(  cox_cont_age_t5_model )[2])
abline(h=0,col=2)
````
